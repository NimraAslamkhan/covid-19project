```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv( r'C:\Users\123\Downloads\read.csv')
print(df.head())
```

```python
columns = df.columns
print(columns)
```

```python
tail_rows = df.tail()

# Display the last few rows
print(tail_rows)
```

```python
data_description = df.describe()
print(data_description)
```

```python
# Check for null values
null_values = df.isnull()

# Check if any null value exists in the DataFrame
any_null = null_values.any().any()

# Display the result
if any_null:
    print("There are null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```

```python
sns.scatterplot(x='New cases', y='Recovered', data=df)

# Optionally, add a title
plt.title('Scatter Plot of New cases vs Recovered')

# Show the plot
plt.show()
```

```python
sns.scatterplot(x='Confirmed last week', y='Recovered', data=df)

# Optionally, add a title
plt.title('Scatter Plot of Confirmed last week vs Recovered')

# Show the plot
plt.show()
```

```python
sns.scatterplot(x='Confirmed', y='Deaths',hue='Recovered', data=df)

# Optionally, add a title
plt.title('Scatter Plot of Confirmed  vs Deaths')

# Show the plot
plt.show()
```

```python
sns.pairplot(df)

# Optionally, add a title
plt.title('Pairplot of the Data')

# Show the plot
plt.show()
```

```python
sns.lineplot(data=df, x='Recovered', y='Deaths', hue='WHO Region')

# Optionally, add a title
plt.title('Line Plot of Deaths Over Time by WHO Region')

# Rotate x-axis labels for better readability if necessary
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
